```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"cancer_data.csv")
df['Level'].unique()
```

```python
# A
plt.figure(figsize=(6,4))
sns.countplot(x='Level', data=df, palette='spring')
plt.title("Distribution of Target Variable (Level)")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()


```

```python
# B
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True, color='purple')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# C
plt.figure(figsize=(6,4))
sns.boxplot(x='Level', y='Obesity', data=df, palette='spring')
plt.title("Obesity by Level")
plt.xlabel("Level")
plt.ylabel("Obesity")
plt.show()
```

```python

# D
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap='spring', fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
