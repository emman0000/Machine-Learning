```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_data.csv")
df['Level'].unique()
```

```python
# Total number of columns
print("Total number of columns:", df.shape[1])

# First 50 records
print("\n--- First 50 records ---")
display(df.head(50))

# Last 30 records
print("\n--- Last 30 records ---")
display(df.tail(30))

# Info (memory usage, dtypes, non-null counts)
print("\n--- Dataset Info ---")
df.info(memory_usage='deep')

# Dimensions (rows x columns)
print("\n--- Shape (rows, columns) ---")
print(df.shape)

# Total size (# of elements = rows * columns)
print("\n--- Size (total elements) ---")
print(df.size)
```

```python
# Number of columns
print("Number of columns:", df.shape[1])

# Total number of samples (rows)
print("Total number of samples:", df.shape[0])
```

```python
print("Number of records for each category in target variable:")
print(df['Level'].value_counts())

# Graph
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='Level', data=df, palette="spring")
plt.title("Number of Records per Category in Target Variable")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python

subset1 = df.loc[550:900, 'Gender':'Coughing of Blood']
subset1.to_csv("subset1.csv", index=False)

# 5. Statistical analysis
print("\n--- Mean ---")
print(subset1.mean(numeric_only=True))

print("\n--- Median ---")
print(subset1.median(numeric_only=True))

print("\n--- Mode ---")
print(subset1.mode().iloc[0])

print("\n--- Standard Deviation ---")
print(subset1.std(numeric_only=True))

print("\n--- Minimum ---")
print(subset1.min(numeric_only=True))

print("\n--- Maximum ---")
print(subset1.max(numeric_only=True))
```
